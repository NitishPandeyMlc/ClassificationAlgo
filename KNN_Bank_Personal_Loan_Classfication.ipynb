```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
bdf=pd.read_csv('Bank_Personal_Loan_Modelling-1.csv')
```

```python
bdf.shape
```

```python
bdf.head()
```

```python
bdf.info()
```

```python
bdf.isnull().sum()
```

```python
bdf=bdf.drop(['ID','ZIP Code'],axis=1)
```

```python
bdf.columns
```

```python
bdf['Personal Loan'].value_counts()
```

```python
X=bdf.drop('Personal Loan',axis=1)
```

```python
y=bdf['Personal Loan']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
sc=StandardScaler()
sc=sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

```python
Knn=KNeighborsClassifier(n_neighbors=3)
Knn.fit(X_train,y_train)
```

```python
Knn.score(X_train,y_train)
```

```python
Knn.score(X_test,y_test)
```

```python
y_pred=Knn.predict(X_test)
```

```python
expt=y_test
```

```python
confusion_matrix(expt,y_pred)
```

```python
print(classification_report(expt,y_pred))
```
